# Variables in PyTorch
To do Machine Learning, we often need to compute the variable's gradient to proceed for optimization. This can be done easily in Pytorch as they have a very powerful functionaility known as automatric differentiation (autograd). 

```python
from torch.autograd import Variable
```

To start with, consider building a simple tensor representing the following equation:
$$y = x^2$$ 
where x = 10 

```python
#Build y = x**2
x = torch.FloatTensor([10])
y = x**2
```

We are now interested into computing the gradient of the above expression at x = 10. Besides doing it by hand, we can use pytorch and do it automatically. First of all, we need to start defining it as a Variable.

```python
#Define the variable
x_var = Variable(x, requires_grad = True)
y_var = x_var**2

#Run back propagation
y_var.backward()

#Now enquire the gradient at of x_var
x_var.grad 
```

Built upon the above idea, we can build a computational graph (image a neural network) using these variables and run backward propagation to optain the gradients automatically! (more on autograd next time)

Every variable instances has two attributes: 
* `.data` contain initial tensor 
* `.grad` contain gradients for the corresponding tensor

However, we should manually zero gradients before calculating new ones.

```python
# Manually cleaning the gradients:
x_var.grad.zero_()
```

# Demonstration: Linear Regression

Linear Regression is the most basic machine learning model that basically fits a best fit lines between data points. Its aim is to minimize the following objective funciton

$$L_{MSE} = \frac{1}{n}\sum_{i=1}^n (y_i - f(x_i))$$

and $f(x_i)$ takes the following form:

$$f(x_i) = w_0 + w_1x_1 + w_2x_2 +... +w_px_p$$


Creating models in pytorch follow a simple framework:
1. Create a class
2. Declare your forward pass
3. Tune the HyperParameters

```python
# The following code define a class for Linear Regression
class LinearRegressionModel(nn.Module):
    
    def __init__(self, input_dim, output_dim):
        
        super(LinearRegressionModel, self).__init__()
        self.linear = nn.Linear(input_dim, output_dim)
        
    def forward(self, x):
        out = self.linear(x)
        return out
input_dim = 1
output_dim = 1
    
```

Now we follow the following steps:
1. Create instance of model
2. Select Loss Criterion
3. Choose Hyper Parameters

```python
#Create instances of model
model = LinearRegressionModel(1,1)

#Select Loss Criterion
criterion = nn.MSELoss()
l_rate = 0.01 #learning rate 
optimiser = torch.optim.SGD(model.parameters(), lr = l_rate)

#Set the number of iteration for optimization
epochs = 2000
```

Now train the model

```python
#Create fake data
x_vals = np.random.rand(50)
x_train = np.asarray(x_vals,dtype=np.float32).reshape(-1,1)
m = 1
alpha = np.random.rand(1)
beta = np.random.rand(1)
y_correct = np.asarray([2*i+m for i in x_vals], dtype=np.float32).reshape(-1,1)
```

#### Short recap on gradient descent
Gradient descent is numerical optimisation technique that approximates the optimal value of a function. It takes the general form

$$w^{new} = w^{old} - \epsilon \frac{dL}{dw}\bigg|_{w = w^{old}}$$

```python
for epoch in range(epochs):
    epoch+=1
    
    inputs = Variable(torch.from_numpy(x_train))
    labels = Variable(torch.from_numpy(y_correct))
    
    #clear grads 
    optimiser.zero_grad()
    
    #pass a forward
    outputs = model.forward(inputs)
    loss = criterion(outputs, labels)
    loss.backward()
    optimiser.step() #update the parameters
    print('epoch {},loss {}'.format(epoch, loss.data[0]))
```

```python
# Printing the Predictions
predicted = model.forward(Variable(torch.from_numpy(x_train))).data.numpy()

plt.plot(x_train, y_correct, 'go', label = 'from data', alpha = .5)
plt.plot(x_train, predicted, label = 'prediction', alpha = 0.5)
plt.legend()
plt.show()
model.state_dict()
```
